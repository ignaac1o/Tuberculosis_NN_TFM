```python
import numpy as np
import cv2
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
photos="DATA_SPLIT_JPG_CSV/photos"
annotations="DATA_SPLIT_JPG_CSV/anotations"
```

```python
gtvalues=[]
count=0

for e,i in enumerate(os.listdir(annotations)):
    if i.startswith("tuberculosis"):
        filename=i.split(".")[0]+".jpg"
        image=cv2.imread(os.path.join(photos,filename))
        df=pd.read_csv(os.path.join(annotations,i))
        number=df.shape[0]
        gtvalues.append(number)

gtvalues=np.array(gtvalues)
```

```python
(unique,counts)=np.unique(gtvalues,return_counts=True)
```

```python
# Create bars
plt.subplot(1,5,(1,3))
plt.bar(unique, counts)

plt.title('Bar Plot')
plt.xlabel('TB Bacillus')
plt.ylabel('Images')

plt.subplot(1,5,(4,5))
plt.boxplot(gtvalues)
plt.ylabel('TB Bacillus')
plt.title('Box Plot')
# Show graphic

plt.subplots_adjust(wspace=2,hspace=0.8)
plt.show()


```

```python
sns.distplot(gtvalues, hist=True, 
             bins=unique, color = 'darkblue', 
             hist_kws={'edgecolor':'black'})

plt.title('Density Plot, number of bacillus per image')
plt.xlabel('TB Bacillus')
plt.ylabel('Density')
```
